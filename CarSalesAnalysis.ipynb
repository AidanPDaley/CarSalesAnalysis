```python
vdf["driven"] = vdf["mileage"] > 0
vdf.groupby("driven")["price"].mean().sort_values(ascending=False)
```

```python
# Exterior Color
def color(c):
    c = str(c).lower()
    if 'white' in c:
        return 'white'
    elif 'black' in c:
        return 'black'
    elif 'red' in c:
        return 'red'
    elif 'grey' in c or 'gray' in c:
        return 'grey'
    elif 'blue' in c:
        return 'blue'
    elif 'silver' in c:
        return 'silver'
    elif 'metallic' in c:
        return 'metallic'
    elif 'green' in c:
        return 'green'
    else:
        return 'other'
       
vdf["carColor"] = vdf['exterior_color'].apply(color)
print(vdf.groupby("carColor")["price"].mean().sort_values(ascending=False))
vdf["carColor"].value_counts().sort_values(ascending=False)
```

```python
colors_dict = {}

def color2(paintColor):
    paintColor = str(paintColor).lower().split()
    for c in paintColor:
        if c in colors_dict:
            colors_dict[c] += 1
        else:
            colors_dict[c] = 1
vdf["exterior_color"].apply(color2)

colors_df = pd.DataFrame.from_dict(colors_dict, orient="index", columns=["count"])
colors_df = colors_df.reset_index().rename(columns={"index": "word"})
colors_df = colors_df.sort_values(by="count", ascending=False) 
colors_df["percent"] = (colors_df["count"] / vdf.shape[0]) * 100
colors_df.head(10)


```

```python
vdf["carColor"].value_counts().plot(kind="bar")
plt.title("Popularity of Car Color")
```
